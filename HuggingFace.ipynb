```python
import transformers
print(transformers.__version__)
```

```python
from transformers import pipeline
```

```python
sentiment = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

sentiment("I love learning Generative AI")
```

```python
qa = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

qa(
    question="What is machine learning?",
    context="Machine learning is a field of AI that allows systems to learn from data."
)
```

```python
generator = pipeline(
    "text-generation",
    model="gpt2"
)

generator(
    "Artificial intelligence will",
    max_length=40,
    num_return_sequences=1
)
```

```python
summarizer = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

text = """
Machine learning is a field of artificial intelligence that focuses on building
systems that can learn from data, identify patterns, and make decisions with minimal
human intervention. It is widely used in recommendation systems, image recognition,
and natural language processing.
"""

summarizer(
    text,
    max_length=60,
    min_length=30,
    do_sample=False
)
```

```python
reviews = [
    "This course is excellent",
    "The explanation was confusing",
    "Very helpful content"
]

sentiment(reviews)
```
